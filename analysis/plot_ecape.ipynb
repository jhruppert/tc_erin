```python
import xarray as xr
import cmocean
import matplotlib.pyplot as plt
from matplotlib import rc
import cartopy.feature as cpf
import cartopy.crs as ccrs
import numpy as np
import datetime
import tropycal.tracks as tracks
```

Read in data

```python
dnc = "/ourdisk/hpc/radclouds/auto_archive_notyet/tape_2copies/robby/erin/ecape/"

# ds17 = xr.open_dataset(f"{dnc}ecape_20070817.nc")
ds18 = xr.open_dataset(f"{dnc}ecape_20070818.nc")
ds19 = xr.open_dataset(f"{dnc}ecape_20070819.nc")
# extract dimensions
lon = ds18.longitude
lat = ds18.latitude
#Import Dataset for Best Track Data
url = "https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2022-042723.txt"
basin = tracks.TrackDataset(basin='north_atlantic', atlantic_url=url)
storm = basin.get_storm(('erin',2007))
storm = storm.to_xarray()
```

Plotting set up

```python
# plotting setup

# Set the font weight for plot titles
plt.rcParams['axes.titleweight'] = 'bold'

# Set the font weight for x-axis and y-axis labels
plt.rcParams['axes.labelweight'] = 'normal'

plt.rcParams['text.latex.preamble'] = r'\usepackage{bm}'

# Other plotting setup
rc('font', family='sans-serif')
# plt.rcParams['font.family'] = 'DejaVu Sans'
rc('font', weight='normal', size=17.5)
rc('figure', facecolor='white') 
```

Plot 08/18

```python
cmap = cmocean.cm.thermal
clevs = np.arange(0,3000.11,100)

for jt in range(ds18.time.size):
    # find time index for storm track data
    try:
        time_idx = np.where(storm.time.values == ds18.time[jt].values)[0][0]
    except IndexError:
        time_idx = np.where(storm.time.values == ds18.time[int(jt-1)].values)[0][0]
        pass

    fig, ax = plt.subplots(figsize=(9,8), subplot_kw={'projection': ccrs.PlateCarree()})

    cf = ax.contourf(lon, lat, ds18.ecape[jt], clevs, cmap=cmap)

    # plot star for storm position at time jt
    ax.scatter(storm.lon[time_idx], storm.lat[time_idx], color='white', marker='*', s=1000, transform=ccrs.PlateCarree(), zorder=2)
    # plot line of full storm track
    plt.plot(storm.lon[3:], storm.lat[3:], linestyle='--', marker='o', color='black', zorder=1)

    cb = plt.colorbar(cf, ax=ax)
    cb.set_label("ECAPE [J kg$^{-1}$]")
    cb.set_ticks(np.arange(0,3001,500))

    ax.coastlines()
    ax.add_feature(cpf.BORDERS)
    ax.add_feature(cpf.STATES)

    dt = (ds18.time[jt].values)
    dt_str = np.datetime_as_string(dt)
    date_str = dt_str[:10]
    hr_str = dt_str[11:13]
    title_str = f"RUC ECAPE valid {date_str} {hr_str}:00 UTC"

    ax.set_title(title_str)

    fig.tight_layout()
    
    dout = f"{dnc}RUC_ECAPE_{dt_str[:13]}.png"
    plt.savefig(dout, dpi=300)
```

Combine output images into gif

```python
import os
import imageio

# Directory containing the PNG files
input_dir = dnc

# Output GIF file path
output_gif = f"{dnc}RUC_ECAPE.gif"

# Get the list of PNG files in the directory
png_files = sorted([f for f in os.listdir(input_dir) if f.endswith('.png')])

# Create a list to store the images
images = []
for png_file in png_files:
    # Read each PNG file and append it to the list of images
    file_path = os.path.join(input_dir, png_file)
    images.append(imageio.imread(file_path))

# Save the list of images as a GIF file
imageio.mimsave(output_gif, images, duration=1000)

print(f"GIF file saved to: {output_gif}")
```
